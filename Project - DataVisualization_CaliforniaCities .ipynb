```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities = pd.read_csv('california_cities.csv')
cities.head(5)
```

```python
sns.set_style('whitegrid')
```

```python
# Extract Latd = Vi do, Longd = kinh do
latd, longd = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']
```

```python
# Plot using Pyplot API
plt.scatter(longd, latd, c = np.log10(population), 
            cmap = 'viridis', 
            s = area, linewidths = 0, alpha = 0.8);
# sns.set_style('darkgrid')
plt.clim(3, 7)
plt.axis('equal');
plt.xlabel('Longlatitube')
plt.ylabel('Latitube')
plt.title('Colafornia Cities: Population and Area Distribution')
plt.colorbar(label = 'log$_{10}$(population)');

# create a lengend for cities' size
area_range = [50, 100, 300, 500] 
for area in area_range:
    plt.scatter([], [], s= area, c = 'k', 
                alpha = 0.4, 
                label = str(area) + 'km$^2$')
plt.legend(labelspacing = 1, title = "City Area")
    
```
